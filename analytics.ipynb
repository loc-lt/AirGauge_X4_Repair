```python
# Giới hạn trên và dưới của đường kính trong con hàng ngắn 
LSL_ChannelB_ID = 11.110
USL_ChannelB_OD = 11.150

# Giới hạn trên và dưới của đường kính ngoài con hàng ngắn
LSL_ChannelA_ID = 15.989
USL_ChannelA_OD = 15.999

# Khi báo Threshold của đường kính trong và đường kính ngoài để biết được khi 
# nào bắt đầu lấy dữ liệu
Threshold_B = 11.16
Threshold_A = 15.980
```

```python
# Đọc lấy dữ liệu của toàn bộ quá trình đo đường kính 1 con hàng BW Mini bằng Airgauge
with open("21122023_1.txt",'r') as filename:
    txt = filename.read()

# Lấy dữ liệu từng dòng
list_total = txt.split('\\r\\n\n')

# Kiểm tra
print(list_total)
```

```python
# Khai báo các danh sách lưu dữ liệu để có thể vẽ biểu đồ
list_a = []
list_b = []
x_axis = []
```

```python
# # Khai báo biến để biết được đã và đang đo đường kính nào của kênh nào
# dict_location_a = {"Begin": None, "Middle": None, "End": None} # vì đo đường kính ngoài nên phải đo 3 chỗ đầu, giữa, cuối -> 0 là đầu, 1 là giữa, 2 là cuối
# dict_location_b = {"Begin": None, "End": None} # vì đo đường kính trong nên phải đo 2 chỗ đầu, cuối -> 0 là đầu, 1 là cuối 

# Các biến khác
balance_var_outside = 11.12
balance_var_inside = 15.995
```

```python
# Đo một con hàng 
check_measure_inside_diameter = False
check_measure_outside_diameter = False

flag_check_B_list = [0, 0]
flag_check_A_list = [0, 0, 0]


for record_idx in range(len(list_total)):
    # In ra nguyên văn phần tử dòng dữ liệu hiện tại
    print(list_total[record_idx])

    data_record = list_total[record_idx].split()

    # In từng phần tử của dòng dữ liệu 
    print(data_record)

    # Thêm dữ liệu và list_a và list_b để theo dõi sự thay đổi đường kính kênh B 
    # (đo đường kính trong của đầu và cuối) và sự thay đổi đường kính kênh C 
    # (đo đường kính ngoài của đầu, giữa, cuối)
    for item_idx in range(len(data_record)):
        # Lưu dữ liệu vào ACH (đo đường kính ngoài)
        if data_record[item_idx] == 'ACH':
            if float(data_record[item_idx+1][:-2]) < 16:
                list_a.append(float(data_record[item_idx+1][:-2]))
                if float(data_record[item_idx+1][:-2]) > Threshold_A:
                    # Thông báo rằng lần lấy dữ liệu này là chưa/đang đo đường kính trong/ngoài
                    if not check_measure_outside_diameter:
                        print("Bắt đầu đo đường kính ngoài ở lần thứ ", record_idx + 1)
                        check_measure_outside_diameter = True
                    else:
                        print("Vẫn đang đo đường kính ngoài ở lần thứ ", record_idx + 1)
                else:
                    if check_measure_outside_diameter:
                        print("Kết thúc đo đường kính ngoài ở lần thứ ", record_idx + 1)
                        check_measure_outside_diameter = False
                    else:
                        print("Chưa đo đường kính ngoài ở lần thứ ", record_idx + 1)
        
        # Lưu dữ liệu vào BCH (đo đường kính trong)
        elif data_record[item_idx] == 'BCH':
            if float(data_record[item_idx+1][:-2]) < 16:
                list_b.append(float(data_record[item_idx+1][:-2]))
                if float(data_record[item_idx+1][:-2]) < Threshold_B:
                    # Thông báo rằng lần lấy dữ liệu này là chưa/đang đo đường kính trong/ngoài
                    if not check_measure_inside_diameter:
                        print("Bắt đầu đo đường kính trong ở lần thứ ", record_idx + 1)
                        check_measure_inside_diameter = True
                    else:
                        print("Vẫn đang đo đường kính trong ở lần thứ ", record_idx + 1)

                    # Xem xét xem có bắt đầu đo đường kính trong/ngoài chưa
                    if flag_check_B_list[0] == 0:
                        if abs(balance_var_outside - float(data_record[item_idx+1][:-2])) < 0.0001:
                            print("Sẵn sàng đo đường kính trong đầu 1")
                        else:
                            print("Bắt đầu đo đường kính trong đầu 1")
                            flag_check_B_list[0] = 1
                    else:
                        if flag_check_B_list[1] == 0:
                            if abs(balance_var_outside - float(data_record[item_idx+1][:-2])) < 0.0001:
                                print("Kết thúc đo đường kính trong đầu 1 -> Sẵn sàng cho việc đo đường kính trong đầu 2")
                                flag_check_B_list[1] = 1
                            else:
                                print("Vẫn đang đo đường kính trong đầu 1")
                        else:
                            # Lúc này đã vào vòng if ngay phía trên để flag_check_B_list[1] = 1
                            if abs(balance_var_outside - float(data_record[item_idx+1][:-2])) < 0.0001:
                                print("Kết thúc đo đường kính trong đầu 2 -> Kết thúc đo đường kính trong của một con hàng")
                                flag_check_B_list = [0, 0]
                            else:
                                print("Vẫn đang đo đường kính trong đầu 2")   
                else:
                    if check_measure_inside_diameter:
                        print("Kết thúc đo đường kính trong ở lần thứ", record_idx + 1)
                        check_measure_inside_diameter = False
                    else:
                        print("Chưa đo đường kính trong ở lần thứ ", record_idx + 1)
```
